# Application : Plot multi-functions

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

# Gaussian Distribution

```python
x = np.arange(-6,6,0.001)

plt.figure(figsize = [10,3])
sigma =1.02
mu = 0.01
fx = 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-(x - mu)**2/(2*sigma**2))
plt.plot(x,fx)

plt.xlabel('x',fontsize =25)
plt.ylabel('probability')
plt.show()
```

```python
xs = np.arange(-15,15,0.001)
plt.figure(figsize = [10,5])

sigmas = [ 1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0]
mus = [0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0]

for s,m in zip(sigmas,mus):
    n = 1/np.sqrt(2*np.pi*s**2)
    fx = n*np.exp(-(xs-m)**2/(2*s**2))
    plt.plot(xs,fx, label = "mu="+str(m)+",sigma="+str(s))
    
plt.xlabel('x')
plt.ylabel('probability')
plt.show()
```

```python
xs = np.arange(-13,13,0.001)
plt.figure(figsize = [13,15])

sigmas = [ 1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0]
mus = [0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0]

k = 1
for s,m in zip(sigmas,mus):
    plt.subplot(3,3,k)
    n = 1/np.sqrt(2*np.pi*s**2)
    fx = n*np.exp(-(xs-m)**2/(2*s**2))
    
    plt.plot(xs,fx, label = "mu="+str(m)+",sigma="+str(s))
    plt.xlabel('x',fontsize =27)
    plt.ylabel('prabability')
    plt.legend()
    k = k+1
plt.show()
```
